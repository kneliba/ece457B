```python
import numpy as np
import matplotlib.pyplot as plt
import time
from IPython import display
```

```python
inputRGB = np.array([
[255,0,0],
[0,255,0],
[0,0,255],
[255,255,0],
[255,0,255],
[0,255,255],
[128,128,0],
[128,0,128],
[0,128,128],
[255,128,0],
[255,0,128],
[128,255,0],
[0,255,128],
[128,0,255],
[0,128,255],
[255,20,147],
[220,20,60],
[255,51,51],
[255,153,51],
[255,255,51],
[51,255,51],
[153,255,51],
[51,255,153],
[51,255,255]])

normRGB = inputRGB/255
print(len(inputRGB))

plt.imshow(np.reshape(normRGB,(normRGB.shape[0],1,3)))
print(normRGB.shape)
```

```python
# Initialize the system
space_size = 100 # 100 x 100 grid of neurons
alpha_0 = 0.8
Nc = 20
max_epochs = 1000

# Initialize random weights
w = np.random.random((space_size,space_size,3))
plt.imshow(w)
```

```python
# Initialize random weights
w = np.random.random((space_size,space_size,3))

epoch = 1
sigma_0 = 1
plot_ind = [1,20,40,100,1000]

while epoch <= max_epochs:
    alpha = alpha_0 * np.exp(-1 * epoch / max_epochs)
    sigma = sigma_0 * np.exp(-1 * epoch / max_epochs)
    
    for x in normRGB:
        diff = np.linalg.norm(x - w, axis =2)
        ind = np.unravel_index(np.argmin(diff, axis=None), diff.shape)

        for i in range(100):
            for j in range(100):
                if i >= 0 and j>=0 and i < space_size and j < space_size:
                    d = np.sqrt((ind[0] - i)**2 + (ind[1] - j)**2)
                    n = np.exp(-1 * d**2 / (2*sigma**2))
                    w[i][j] += alpha * n * (x-w[i][j])

    if epoch in plot_ind:
        print("Epoch Number: {}".format(epoch))
        plt.imshow(w)
        display.display(plt.gcf())
    
    
    epoch += 1
```
